```python
def is_question(text):
  return "?" in text or "요" in text[-3:]
```

```python
text = "안녕하세요, 롯데월드는 어떻게 가나요?"

is_question(text)
```

```python
# 에러 방지용 예외처리, try: except

try: # try에는 에러가 날 수 있는 구문을 넣어줌.
    result = 10 / 0 # 에러가 안나면 try 구문 안에 있는 내용이 실행
except:
    print("에러발생") # 에러가 날 경우 except 구문에 넣어준 내용을 실행
```
