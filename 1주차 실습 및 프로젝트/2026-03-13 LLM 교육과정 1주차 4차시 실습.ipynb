```python
from langchain_core.messages import SystemMessage, HumanMessage

# 모델의 역할(페르소나) 정의
persona = "MLOps 리더"
# 사용자 질문 정의
question = "Langchain 체인을 배포할 때, 고려해야 하는 리스크 두가지를 알려주세요."

# 메시지 리스트 생성: 시스템 설정과 사용자 질문을 포함
messages = [
    # SystemMessage: 모델에게 페르소나 및 답변 지침 부여
    SystemMessage(content = f"당신은{persona}입니다. 실무 팁 위주로 두 문장 이하로 답변해주세요."),
    # HumanMessage: 실제 사용자가 전달하는 입력 메시지
    HumanMessage(content = question)
]
```

```python
def ask_expert(persona, question):

  messages = [
      SystemMessage(content = f"당신은{persona}입니다. 실무 팁 위주로 두 문장 이하로 답변해주세요."),
      HumanMessage(content = question)
  ]
  response = llm.invoke(messages)

  return response.content, response.response_metadata['token_usage']
```

   - `ChatOpenAI` 클래스를 통해 모델 종류와 API 키를 설정합니다.

3. **LangChain Core 메시지 구조**
   - `SystemMessage`: AI의 페르소나(역할)와 행동 지침을 설정합니다.
   - `HumanMessage`: 사용자가 전달하는 실제 질문이나 요청 사항입니다.
   - `AIMessage`: AI 모델이 생성하여 반환한 답변 객체입니다.

4. **주요 메서드 및 속성**
   - `llm.invoke(messages)`: 설정된 메시지 리스트를 전달하여 ⭐ 답변을 생성하는 핵심 메서드입니다.
   - `response.content`: 생성된 답변의 텍스트 내용만 추출합니다.
   - `response.usage_metadata`: 토큰 사용량 등 실행 관련 메타데이터를 확인합니다.

### invoke, stream, batch

```python
# 자주 쓰이는 메서드
# invoke: 답변을 내주는것
# stream: 답변을 한꺼번에 주지 않고 실시간으로 생성되는 대로 끊어서 보여주는 방식.
# batch: 여러 개의 질문을 리스트로 묶어 한꺼번에 효율적으로 처리하는 방식.

from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

llm = ChatOpenAI(model = 'gpt-4o-mini', openai_api_key=api_key, temperature = 0.7)

messages = [
    SystemMessage(content = f"당신은 친절한 AI 튜터입니다. 간결하게 답변하세요."),
    HumanMessage(content = "Langchain이란 무엇인가요?")
]
llm.invoke(messages)
```

```python
# stream

# '조각(chunk)'이란 모델이 답변을 생성할 때 한꺼번에 모든 문장을 만드는 것이 아니라
# 단어나 글자 단위로 조금씩 생성해서 보내주는 데이터를 의미
for chunk in llm.stream("머신러닝의 3가지 유형을 나열해줘"):
  print(chunk.content, end="", flush=True)
```

```python
# batch

questions = [
    "파이썬의 장점 하나만 말해줘.",
    "자바스크립트의 장점 하나만 말해줘",
    "러스트의 장점 하나만 말해줘"
]

results = llm.batch(questions)
```

```python
results
```

```python
for answer in results:
  print(answer.content)
  print('\n')
```

### PromptTemplate

- **PromptTemplate**: 사용자의 입력값(변수)을 미리 정의된 템플릿 문자열에 삽입하여 완성된 프롬프트를 만드는 도구입니다.

```python
# 하드코딩 예시
llm.invoke('당신은 파이썬 전문가입니다. 함수는 어떻게 사용하나요?')
llm.invoke('당신은 자바 전문가입니다. 함수는 어떻게 사용하나요?')
```

```python
# 랭체인을 안 쓸 경우 템플릿 함수를 직접 만든 예시

def chat_prompt_template(role, style, question):
  system_prompt = f"당신은 {role} 전문가입니다. {style}로 답변하세요."
  messages = [
      {'role':"system", "content": system_prompt},
      {'role':"user", "content": question}

  ]
  return messages
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 {role} 전문가입니다. {style}(로) 답변하세요"), # f-string 안써도 알아서 작업 됨
    ("human", "{question}")
])
```

```python
prompt
```

```python
prompt.input_variables # ChatPromptTemplate.from_messages로 만든 객체 안에 {question}, {role}, {style}이 있다는 뜻
```

```python
results = prompt.invoke({ # llm.invoke() 대신
    "role":"파이썬",
    "style": "간결하게",
    "question": "파이썬에서 함수 쓰는 방법?"
})
```

```python
results
```

```python
for msg in results.messages:
  print(msg.content)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 {role} 전문가입니다. 핵심만 3줄로 답변하세요"), # f-string 안써도 알아서 작업 됨
    ("human", "{question}")
])

partial_prompt = prompt.partial(role="파이썬")
```

```python
# partial_prompt는 이미 {role}을 채워놓음
prompt.input_variables,partial_prompt.input_variables
```

```python
"""
아래 코드는 랭체인의 핵심 문법인 LCEL(LangChain Expression Language)과 Partial Prompt 기능을 사용한 것.

1. | (파이프 연산자): 앞의 결과를 뒤로 넘겨주라는 '연결 고리'입니다
  - partial_prompt | llm은 "프롬프트에서 만들어진 메시지 뭉치를 바로 LLM(모델)에게 던져줘!"라는 의미.
2. .invoke({...}): "question": "..." } 이라는 입력이 들어가면 아래 단계를 거침
  1): partial_prompt가 입력을 받아 빈칸({question})을 채워 '완성된 메시지'를 만듭니다.
  2): |를 타고 '완성된 메시지'가 llm에게 자동 전달.
  3): llm 객체는 그 메시지를 읽고 답변(response)을 내놓습니다.

"""

response = (partial_prompt | llm).invoke({"question": "함수 어떻게 쓰나요?"})

```

### StrOutputParser JsonOutputParser PydanticOutputParser

- **StrOutputParser**: LLM의 응답(AIMessage)에서 복잡한 메타데이터(토큰 사용량, 모델 정보 등)를 제거하고, 사용자가 실제로 읽는 순수 텍스트(content)만 추출하여 반환해주는 출력 파서입니다.
- StrOutputParser: 모델의 응답 객체에서 메타데이터를 제외하고 순수하게 '문자열(텍스트)'만 뽑아줍니다. 가장 기본적이고 자주 쓰입니다.
JsonOutputParser: 응답 결과물을 Python의 딕셔너리나 JSON 객체 형태로 변환해줍니다. 구조화된 데이터가 필요할 때 사용합니다.
- PydanticOutputParser: 미리 정의한 Pydantic 모델(클래스) 형식에 맞춰 응답을 검증하고 객체로 변환해줍니다. 가장 엄격하고 정교한 구조가 필요할 때 유용합니다.

```python
# Pydantic을 사용하는 이유는 LLM의 답변 형식을 아주 엄격하게 규격화하기 위함
"""
Pydantic으로 클래스를 미리 정의(BaseModel)해두면:

- 자동 검증: 답변이 설정한 타입(문자열, 숫자 등)에 맞는지 확인
- 자동 변환: 모델의 답변을 파이썬 객체로 바로 변환
- 가이드 제공: 어떤 데이터가 필요한지 모델에게 더 명확하게 알려주는 설계도 역할
"""

from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field

class Cityinfo(BaseModel):
  name : str = Field(description="도시이름")
  country : str = Field(description="국가")
  population: str = Field(description = "인구 (대략적)")
  famous_for : list[str] = Field(description = "유명한 것 3가지")


parser = JsonOutputParser(pydantic_object=Cityinfo) # Json으로 답을 주되, 설정하는 방식
# 모델의 '답변'을 JSON(파이썬 딕셔너리(CityInfo 형식)) 형태로 받아내기 위한 도구
```

```python
parser.get_format_instructions()
```

```python
# PydanticOutputParser 특정 수치의 범위를 제한하여 출력할 경우

from langchain_core.output_parsers import PydanticOutputParser

class BookReview(BaseModel):
  title : str = Field(description = "책 제목")
  author : str = Field(description = "저자")
  rating : str = Field(description = "평점 (1-5)", ge=1, le=5)
  summary : str = Field(description = "한줄 요약")
  recommended : bool = Field(description = "추천 여부")

parser = PydanticOutputParser(pydantic_object= BookReview)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "도서 리뷰를 작성해주세요.\n{format_instructions}"),
    ("human", "{book} 책에 대한 리뷰를 작성해주세요")

])

chain = prompt | llm | parser
result = chain.invoke({
    "book" : "장발장",
    "format_instructions": parser

})
```

### 🚀 랭체인 핵심 요약 및 암기 노트 (심화)

#### 1. 실행 메서드 3총사
- **invoke**: `chain.invoke({"topic": "AI"})` → 단일 답변 생성
- **stream**: `for chunk in chain.stream(...): print(chunk)` → 실시간 조각 출력
- **batch**: `chain.batch([{"t": "A"}, {"t": "B"}])` → 병렬 처리

#### 2. 프롬프트 구성 (Prompt Management)
- **ChatPromptTemplate**: `("system", "지침"), ("human", "질문")` 구조로 대화 맥락 유지
- **partial**: `prompt.partial(role="바리스타")` → 페르소나 같은 고정 역할을 미리 주입하여 재사용
- **bind**: `llm.bind(max_tokens=100, stop=["\n"])` → 모델 호출 파라미터 고정

#### 3. LCEL (연결의 핵심)
- **구조**: `프롬프트 | 모델 | 파서`
- **원리**: 앞 단계의 출력이 다음 단계의 입력으로 자동 전달 (파이프 연산자 `|` 사용)

#### 4. 출력 파서 (Output Parsers)
- **StrOutputParser**: `AIMessage`에서 문자열만 쏙 뽑기
- **JsonOutputParser**: Pydantic과 결합하여 `{"key": "value"}` 형태의 구조화된 데이터 강제
- **PydanticOutputParser**: `parser.get_format_instructions()`를 프롬프트에 포함시켜 모델에게 출력 규격을 상세히 가이드

- Q: 유효성 검사가 뭘까
- **유효성 검사(Validation)**란 모델이 내놓은 답변이 우리가 정한 '규칙'에 맞는지 확인하는 절차입니다.

- 예를 들어, rating은 1에서 5 사이의 숫자여야 하고, title은 반드시 문자열이어야 한다고 규칙을 세웠을 때, LLM이 실수로 글자를 넣거나 항목을 빼먹으면 Pydantic이 이를 잡아내어 에러를 발생시키거나 수정을 요청합니다.

- 즉, AI가 가끔 엉뚱한 형식으로 답하는 것을 방지하고, 데이터의 정확성과 신뢰성을 확보하는 안전장치라고 이해하시면 됩니다.
