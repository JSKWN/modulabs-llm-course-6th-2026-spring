```python
import os
from google.colab import userdata # Colab 비밀번호 관리자 가져오기
from openai import OpenAI
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 로컬 환경에서 활용할 경우, `.env' 파일 내에 API 키 명시한 후 불러와야함.
# from dotenv import load_dotenv # .env 파일에 API 키를 명시해두고 불러오는 방식임
# load_dotenv()

# 1. API 키 설정 (Secrets에 저장한 키 불러오기)
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

# 2. 클라이언트 및 모델 초기화
client = OpenAI() # `.env` 파일을 활용할 경우, () 안에 아래 값을 인자로 주면 됨
                  #  api_key = os.getenv("OPENAI_API_KEY")
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

print("✅ 환경 설정 완료")
print(f"모델: gpt-4o-mini")
print(f"API 키 설정: {'✅' if os.environ.get('OPENAI_API_KEY') else '❌'}")
```

```python
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = [
        {
         "role": "user",
         "content": "안녕하세요, 자기소개 부탁드립니다."
        }
    ]
)
```
